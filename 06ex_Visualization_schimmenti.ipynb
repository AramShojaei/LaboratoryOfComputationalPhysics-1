```python
import numpy as np
import matplotlib.pyplot as plt
import math
import scipy.stats as stats
import pandas as pd
import seaborn as sns
import scipy
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
#Esercizio 1
N=2000
mean = 1.5
std = 0.05
n_bins = int(2.5*math.log(N))
x = np.random.normal(mean, std, N)
freq,bin_edges,_ = plt.hist(x,bins=n_bins)
bin_centers = 0.5*(bin_edges[1:] + bin_edges[:-1])
plt.errorbar(bin_centers, freq, yerr=np.sqrt(freq), fmt='none')
ax = plt.gca()
ax.set_ylabel('Frequency')
kde_std = 1.06*x.std()*(x.size)**(-1/5)
#t = np.linspace(x.min(),x.max(),len(x))
g = np.zeros((N,len(bin_centers)))
for i in range(N):
    g[i,:] = stats.norm.pdf(bin_centers, loc=x[i], scale=kde_std)
    #plt.plot(g[i,:])
g = np.sum(g,axis=0)
g = g*np.sum(freq*(bin_edges[1:] - bin_edges[:-1]))/ scipy.integrate.trapz(y=g, x=bin_centers)
plt.plot(bin_centers,g)
plt.show()
```

```python
#Esercizio 2
m1,m2,s1,s2=2,4,0.8,0.5
def sample_cat_gaussian(length):
    x1 = np.random.randn(length,length)*s1+m1
    x2 = np.random.randn(length,length)*s2+m2
    return np.array([x1,x2])

def sample_multicat_gaussian(ncat,length):
    x = np.random.randn(ncat, length, length)
    for i in range(ncat):
        m = np.random.rand()-0.5
        std = np.random.rand()
        x[i,:,:] = x[i,:,:]*std*np.ones((length,length))+ m*np.ones((length,length))
    return x

data = sample_multicat_gaussian(2,200)
plt.scatter(data[0,0], data[0,1])
plt.scatter(data[1,0], data[1,1])
plt.show()


```

```python
data = np.load('residuals_261.npy').item()
df = pd.DataFrame(data)
df = df[np.abs(df['residuals'])<2]
```

```python
a,b,rv,pv,stderr=stats.linregress(df)
```

```python
_,bin_edges = np.histogram(df['distances'], bins=12, range=(0,20))
x = 0.5*(bin_edges[1:] + bin_edges[:-1])
y = np.zeros(x.size)
yerr = np.zeros(x.size)

prevx = 0
for i in range(x.size):
    themask = (df['distances']<x[i]) & (df['distances']>=prevx)
    temp = df[themask]['residuals']
    y[i] = temp.values.mean()
    yerr = temp.values.std()
    prevx = x[i]
sns.jointplot("distances","residuals",data=df, kind='reg',line_kws={'label':"y={0:.1f}x+{1:.1f}".format(a,b),'color':sns.xkcd_rgb['green']})
plt.plot(x,y,color='r')
plt.errorbar(x,y,yerr,color='r',fmt='none')
plt.legend()
plt.show()
```
