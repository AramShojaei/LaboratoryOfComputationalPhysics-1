```python
import csv
import numpy as np
import matplotlib.pyplot as pl
import pandas as pd
```

```python
data = pd.read_csv('data_000637.txt')
N  = len(data)
```

```python
#use for viewing data
pd.set_option('precision', 15)

x = data.describe()['BX_COUNTER']['max']
print(x)
#actual timing
data['tns']=data['TDC_MEAS']*(25/30)+data['BX_COUNTER']*25+data['ORBIT_CNT']*(25*x)
#data['tns']=data['tns']-data['tns'][0]
print('Elapsed time for data taking:', data['tns'][N-1]-data['tns'][0], 'nanoseconds')
data
```

```python

heads = np.zeros(N,dtype=int)
for i in range(N):
    heads[i] = np.random.randint(0,2)
data['HEAD'] = heads
data
```
